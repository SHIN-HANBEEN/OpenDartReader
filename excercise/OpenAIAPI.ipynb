```python
# 사전 설정
from openai import OpenAI
import openai
import os
```

```python
# OpenAI API 키 설정 (환경 변수에서 가져오기)
api_key = os.getenv('OPENAI_API_KEY')
openai.api_key = api_key
client = OpenAI()
```

```python
# API 호출
completion = client.chat.completions.create(model="gpt-5-nano",
                            messages=[{'role':'user','content':'인공지능에 대해서 설명해줘'}] )

print(completion)
print('=============================')
print( completion.choices[0].message.content)
print( completion.usage.total_tokens)
```

### 각 role 의 역활
- system role : 역할 부여
- user role : 사용자 질문( 프롬프트)
- assistant role: gpt 답변(컴플리션)

```python
completion = client.chat.completions.create(model="gpt-5-nano",
    messages=[
        {'role':'system', 'content':'너는 전문 영어 번역가야. 아래 문장을 자연스럽게 영어로 번역해줘'},
        {'role':'user',
            'content':'요즘 날씨가 참 좋네'}
        ])

print( completion )
```
